```python
#Xử lý học phần thay thế
import pandas as pd

# Đổi kiểu dữ liệu của cột Ghi chú thành string
df['Ghi chú'] = df['Ghi chú'].astype(str)

# Danh sách mã học phần cần cập nhật
merge_mapping = {
    'SMT1003': 'Mã học phần thay thế SMT1003',
    'SMT1002': 'Mã học phần thay thế SMT1002',
    'ACC1001': 'Mã học phần thay thế ACC1001',
    'LAW3034': 'Mã học phần thay thế LAW3034',
    'FIN5009U': 'Mã học phần thay thế FIN5009U',
    'HOS3006': 'Mã học phần thay thế HOS3006',
    'SMT1001': 'Mã học phần thay thế SMT1001',
    'STA2003': 'Mã học phần thay thế STA2003',
    'STA2005': 'Mã học phần thay thế STA2005',
    'ENG2014': 'Mã học phần thay thế ENG2014',
    'TOU3019': 'Mã học phần thay thế TOU3019',
    'ENG2012': 'Mã học phần thay thế ENG2012',
    'ACC1902': 'Mã học phần thay thế ACC1002',
    'ENG1012': 'Mã học phần thay thế ENGELE1'
}

# Hàm cập nhật cột Ghi chú
def update_notes(row):
    if row['Mã học phần'] in merge_mapping:
        return merge_mapping[row['Mã học phần']]
    else:
        return row['Ghi chú']

# Áp dụng hàm cho cột Ghi chú
df['Ghi chú'] = df.apply(update_notes, axis=1)

# In ra kết quả
print(df)
```

```python
import numpy
import pandas as pd
import math

def assign_exam_schedule(df, max_shifts, max_rooms, students_per_room):
    # Bước 1: Tính bậc cho mỗi học phần
    #Sử dụng nunique để đếm số lượng MSV duy nhất cho mỗi mã học phần, sử dụng reset_index để tạo 1 cột lưu trữ chỉ số cũ
    degree_df = df.groupby('Mã học phần')['MSV'].nunique().reset_index(name='Bậc')

    # Tạo danh sách V là các học phần có bậc lớn nhất sắp xếp từ lớn đến bé
    sorted_nodes = degree_df.sort_values(by='Bậc', ascending=False)['Mã học phần'].tolist()     
    #Tạo các cạnh nối giữa các mã học phần
    edges = set()

    # Nhóm các sinh viên dựa trên sinh viên đăng ký mã học phần đó 
    for _, group in df.groupby('MSV'): 
        #Tạo ra danh sách chứa các mã học phần trong nhóm sinh viên hiện tại 
        
        subjects = group['Mã học phần'].tolist()
        
        #Tạo ra các cặp cạnh i,j trong danh sách subject. Các cặp này sẽ tạo thành các cạnh biễu diễn bằng giá trị i,j
        edges.update([(subjects[i], subjects[j]) for i in range(len(subjects)) for j in range(i+1, len(subjects))])
    
    course_students = df.groupby('Mã học phần mới')['MSV'].count()
    course_rooms = dict([[kcs, math.ceil(course_students[kcs]/students_per_room)] for kcs in course_students.keys()])

    print(course_rooms)

    # Bước 2: Gán i=1 và tô màu i cho học phần đầu tiên trong danh sách V
    students_per_shift = students_per_room*max_rooms
    shifts = numpy.zeros(max_shifts)
    i = 0
    #colored_nodes = []
    paired_shitfs = []  
    df['Ca thi'] = numpy.nan
    print(sorted_nodes)
    while sorted_nodes and i <= max_shifts:
        # Chọn đỉnh đầu tiên trong danh sách V
        start_node = sorted_nodes[0]
        
        # Tô màu i cho đỉnh đang xét
        #unassigned_students = df.loc[df['Mã học phần mới'] == start_node, 'Ca thi']
        no_students = len(df.loc[df['Mã học phần'] == start_node, 'Ca thi'])
        #no_students = course_students[start_node]
        if no_students > students_per_shift:
            if shifts[i] == 0:
                no_required_shifts = math.ceil(no_students/students_per_shift)
                no_node_students_per_shift = math.ceil(no_students/no_required_shifts)
                j = 0
                for k,r in df.loc[df['Mã học phần'] == start_node].iterrows():
                    df.loc[k,'Ca thi'] = i+int(j/no_node_students_per_shift)
                    j += 1
                #for j in range(no_required_shifts):
                #    df.loc[df['Mã học phần mới'] == start_node, 'Ca thi'][j*no_node_students_per_shift:(j+1)*no_node_students_per_shift] = current_shift
                    #for k in range(no_node_students_per_shift):
                    #    df.loc[df['Mã học phần mới'] == start_node, 'Ca thi'][j*no_node_students_per_shift+k] = current_shift
                    #     current_shift += 1

                for j in range(no_required_shifts):
                    shifts[i+j] += math.ceil(no_node_students_per_shift/students_per_room)
                    
                sorted_nodes = sorted_nodes[1:]
                #df.loc[df['Mã học phần mới'] == start_node, 'Ca thi'] = unassigned_students
                #uncolored_nodes.remove(start_node)
        #else:
        #   #Gán màu i cho cột Ca thi, df.loc để truy cập vào 1 hàng cụ thể trong df
        #    df.loc[df['Mã học phần mới'] == start_node, 'Ca thi'] = i
        #    shifts[i] += math.ceil(len(df.loc[df['Mã học phần mới'] == start_node, 'Ca thi'])/students_per_room)
        #    sorted_nodes = sorted_nodes[1:]
        
        #colored_nodes +=[start_node]
        # Kiểm tra xem có mã học phần nào không kề đỉnh () với học phần đang xét thì tô màu cho nó
        #Tạo ra danh sách các đỉnh chưa tô màu có trong danh sách V mà chưa tô màu i
        
        #uncolored_nodes = set([node for node in sorted_nodes if node in df[df['Ca thi'] != numpy.nan]['Mã học phần mới'].tolist()])

        jn = 0
        while jn < len(sorted_nodes):
            node =  sorted_nodes[jn]
            #Danh sách các đỉnh đã được tô màu i
            colored_nodes = set(df[df['Ca thi'] == i]['Mã học phần mới'].tolist())
            if all((node, colored_node) not in edges and (colored_node, node) not in edges for colored_node in colored_nodes):
                # Nếu không kề với các học phần đã tô màu i, tô màu i cho đỉnh đó
                remaining_room=max_rooms-shifts[i]
                if course_rooms[node] <= remaining_room:                
                    df.loc[df['Mã học phần mới'] == node, 'Ca thi'] = i
                    shifts[i] += course_rooms[node]
                    #uncolored_nodes.remove(node)
                    sorted_nodes = sorted_nodes[:jn] + sorted_nodes[jn+1:]
                    #colored_nodes += [node]
                    jn -= 1
            else:
                print(node)
            jn += 1
                
                
                
        # Bước 3: Loại khỏi danh sách V các học phần đã gán màu
        #sorted_nodes = [node for node in sorted_nodes if node in df[df['Ca thi'] != numpy.nan]['Mã học phần mới'].tolist()]

        # Sắp xếp lại danh sách V theo thứ tự bậc giảm dần
        #degree_df = degree_df[degree_df['Mã học phần mới'].isin(sorted_nodes)].sort_values(by='Bậc', ascending=False)

        

        # Tăng giá trị màu i lên 1 đơn vị
        i += 1

    # In ra DataFrame với thông tin ca thi
    result_df = df[['MSV', 'Mã học phần', 'Ca thi']]
    print(result_df)


# Gọi hàm để gán ca thi
MAX_SHIFTS = 100
MAX_ROOMS = 35
STUDENTS_PER_ROOM = 40
assign_exam_schedule(df, MAX_SHIFTS, MAX_ROOMS, STUDENTS_PER_ROOM)
```
